```python
# train has sale price and id and test has only ID  so we remove them and make a copy
df_train_cp=df_train.drop(["SalePrice","Id"],axis=1).copy()
df_test_cp=df_test.drop("Id",axis=1).copy()
```

```python
df_test_cp.shape
```

```python
# combine train and test for one hot Encoding

combined_data =pd.concat([df_train_cp,df_test_cp])
```

```python
combined_data.shape
```

```python
# one hot encoding 
combined_data=pd.get_dummies(combined_data,drop_first=True)
```

```python
combined_data.shape
```

```python
x_train = combined_data.iloc[:1460,:]
x_test=combined_data.iloc[1460:, :]
y_train= df_train["SalePrice"]
```

### Hyper parameter Optimization

```python
booster=['gbtree','gblinear']
learning_rate=[0.05,0.1,0.15,0.20]
max_depth = [2, 3, 5, 10, 15]
min_child_weight=[1,2,3,4]
n_estimators = [100, 500, 900, 1100, 1500]
base_score=[0.25,0.5,0.75,1]

# Define the grid of hyperparameters to search
hyperparameter_grid = {
    'booster':booster,
    'learning_rate':learning_rate,
    'max_depth':max_depth,
    'min_child_weight':min_child_weight,
    'n_estimators': n_estimators,
    'base_score':base_score
    }
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# random search with 4-fold cross validation
random_cv = RandomizedSearchCV(estimator=regressor,
            param_distributions=hyperparameter_grid,
            cv=5, n_iter=50,
            scoring = 'neg_mean_absolute_error',n_jobs = 4,
            verbose = 5, 
            return_train_score = True,
            random_state=42)
```

```python
random_cv.fit(x_train,y_train)
```

```python
random_cv.best_estimator_
```

```python
random_cv.best_params_
```

Final Model (XG Boost)

```python
# train the model again with tunned hyperparameters
regressor1 = XGBRegressor(missing=-999, base_score=0.25, booster='gbtree', colsample_bylevel=1,
          colsample_bynode=1, colsample_bytree=1, gamma=0,
          importance_type='gain', learning_rate=0.1, max_delta_step=0,
          max_depth=2, min_child_weight=1, n_estimators=900,
          n_jobs=1, nthread=None, objective='reg:linear', random_state=0,
          reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
          silent=None, subsample=1, verbosity=1)
regressor1.fit(x_train, y_train)
```

```python
y_pred1 = regressor1.predict(x_test)
print(y_pred1)
```
