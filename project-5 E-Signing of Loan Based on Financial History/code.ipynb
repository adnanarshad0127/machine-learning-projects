```python
dataset=pd.get_dummies(data=dataset,drop_first=True)
```

```python
dataset.shape
```

Restructure the dataset / feature construction

```python
dataset.head()
```

```python
# combine two columns "personal_account_m" and "personal_account_y"
dataset["employment months"]=(dataset.months_employed + (dataset.years_employed*12))
```

```python
dataset.head()
```

```python
dataset=dataset.drop(columns=["months_employed","years_employed"])
```

```python
dataset.shape
```

```python
# combine 2 columns"personal account year " and"personal account month "
dataset['personal_account_month']=(dataset.personal_account_m + (dataset.personal_account_y))
```

```python
dataset=dataset.drop(columns=["personal_account_y","personal_account_m"])
```

```python
dataset.shape
```

```python
dataset.columns
```

split the dataset into train and test

```python
x=dataset.drop(columns=["entry_id","e_signed"])
```

```python
y=dataset["e_signed"]
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test ,y_train,y_test=train_test_split(x,y,test_size=0.2 ,random_state=0)
```

feature Selection

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

building the model

```python
from sklearn.linear_model import LogisticRegression
classifier_lr=LogisticRegression(random_state=0)
classifier_lr.fit(x_train,y_train)
```

```python
y_pred=classifier_lr.predict(x_test)
```

```python
y_pred
```

Import libraries to check accuracy of model

```python
from sklearn.metrics import accuracy_score , confusion_matrix , recall_score ,r2_score , precision_score ,f1_score
```

```python
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

results = pd.DataFrame([['Logistic Regression', acc, prec, rec, f1]],
               columns = ['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
results
```

```python
from sklearn.svm import SVC
classifier_svm = SVC(kernel = 'linear', random_state = 0)
classifier_svm.fit(x_train, y_train)
```

```python
y_pred = classifier_svm.predict(x_test)
```

```python
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

model_results = pd.DataFrame([['SVM_linear', acc, prec, rec, f1]],
               columns = ['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])

results = results.append(model_results, ignore_index = True)
results
```

Model 3 SVM (rbf)

```python
from sklearn.svm import SVC
classifier_svm_2 = SVC(kernel = 'rbf', random_state = 0)
classifier_svm_2.fit(x_train, y_train)
```

```python
y_pred = classifier_svm_2.predict(x_test)

acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

model_results = pd.DataFrame([['SVM_rbf', acc, prec, rec, f1]],
               columns = ['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
```

```python
results = results.append(model_results, ignore_index = True)
results
```

Model 4 Random forest

```python
from sklearn.ensemble import RandomForestClassifier
classifier_rf=RandomForestClassifier(random_state=0 , n_estimators=100, criterion="entropy")

classifier_rf.fit(x_train , y_train)
```

```python
y_pred = classifier_rf.predict(x_test)
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

model_results = pd.DataFrame([['Random Forest', acc, prec, rec, f1]],
               columns = ['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])


results = results.append(model_results, ignore_index = True)
results
```

XGBoost

```python
from xgboost import XGBRFClassifier
classifier_xgb=XGBRFClassifier(random_state=0)
classifier_xgb.fit(x_train,y_train)
```

Final XGBoost

```python
from xgboost import XGBClassifier
classifier = XGBClassifier(missing=-999, base_score=0.25, booster='gbtree', colsample_bylevel=1,
          colsample_bynode=1, colsample_bytree=1, gamma=0,
          importance_type='gain', learning_rate=0.1, max_delta_step=0,
          max_depth=2, min_child_weight=1, n_estimators=900,
          n_jobs=1, nthread=None, objective='reg:linear', random_state=0,
          reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
          silent=None, subsample=1, verbosity=1)

classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)

acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

final_results = pd.DataFrame([['final XGBoost', acc, prec, rec, f1]],
               columns = ['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])


final_results
```

```python
results = results.append(final_results, ignore_index = True)
results
```

k-fold cross validation and confusion matrix

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier, X=x_train, y=y_train, cv=10)

print("Accuracy is {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation is {:.2f} %".format(accuracies.std()*100))
```
